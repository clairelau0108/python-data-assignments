```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas
import csv
import numpy as np
```

```python
def getListPage(pageUrl):
    try:
        res = requests.get(pageUrl)
        mypage = BeautifulSoup(res.text)
    except:
        return

    for session in browser.find_elements_by_css_selector('div.content'):
        article = {}
        try:
            title = session.find_element_by_css_selector('a.title')
            abstract = session.find_element_by_css_selector('p.abstract')
            nickname = session.find_element_by_css_selector('a.nickname')

            article['headline'] = title.text
            article['nickname'] = nickname.text
            article['abstract'] = abstract.text
            article['website'] = title.get_attribute('href')
            articles.append(article)
        except:
            print('error:label')
```

```python
def getNewsDetail(newsUrl):
    res = requests.get(newsUrl)
    mypage = BeautifulSoup(res.text)
    
    article_content = {}
    try:
        times = browser.find_element_by_css_selector('span.publish-time')
        words = browser.find_element_by_css_selector('span.wordage')
        views = browser.find_element_by_css_selector('span.views-count')
        comments = browser.find_element_by_css_selector('span.comments-count')
        likes = browser.find_element_by_css_selector('span.likes-count')

        article_content['time'] = times.text
        article_content['word'] = words.text
        article_content['view'] = views.text
        article_content['like'] = likes.text
    except:
        print('error:content')

    return article_content
```

```python
articles = []
firstPageUrl='https://www.jianshu.com/c/f6b4ca4bb891'
browser = webdriver.Chrome()
browser.get(firstPageUrl)
getListPage(firstPageUrl)

for i in range(2,15):
    listPageUrl='https://www.jianshu.com/c/f6b4ca4bb891?order_by=added_at&page={}'.format(i)
    browser.get(listPageUrl)
    getListPage(listPageUrl)
    
for i in range(0,len(articles)):
    url = articles[i]['website']
    browser.get(url)
    articles[i].update(getNewsDetail(url))
```

```python
total_articles = []
for i in range(0,len(articles)):
    single_article = articles[i].values()
    total_articles.append(single_article)
```

```python
import csv
with open('jianshu.csv','w',encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    header = ['headline','nickname', 'abstract', 'website', 'time', 'word','view','like']
    writer.writerow(header)
    writer.writerows(total_articles)   
```
