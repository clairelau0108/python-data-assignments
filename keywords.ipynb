```python
def split_words(file):
    f = open(file, "r+")
    content =f.read()
    list = str.split(content)
    return list
```

```python
f1 = split_words("trade-wars-news1.txt")
f2 = split_words("trade-wars-news2.txt")
f3 = split_words("trade-wars-news3.txt")
f4 = split_words("trade-wars-news4.txt")
f5 = split_words("trade-wars-news5.txt")
```

```python
totalwords = f1+f2+f3+f4+f5
```

```python
wordcount = {}
for word in totalwords:
    if word in wordcount:
        wordcount[word]=wordcount[word]+1
    else:
        wordcount[word]=1
```

```python
wordcount
```

```python
with open("stopwords.txt") as f:
    stopwords = f.read().split()
```

```python
removedstop = wordcount
for word in stopwords:
    if word in removedstop:
        del removedstop[word]
```

```python
removedstop
```

```python
removedstop = sorted(removedstop.items(), key=lambda d:d[1], reverse = True)
```

```python
removedstop
```

```python
results = removedstop[0:15]
```

```python
results
```

```python
import csv
with open ('keywords.csv','w') as k:
    writer = csv.writer(k)
    header = ['keyword','frequency']
    writer.writerow(header)
    writer.writerows(results)
```
